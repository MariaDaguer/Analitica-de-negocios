0. Pre-operaci√≥n

```python
import numpy as np
import pandas as pd


from  sklearn.tree import  DecisionTreeClassifier
from sklearn.metrics import confusion_matrix


```

1. Base de datos y variable

```python
nxl='/content/2. TechSalesReport_Int_L.xlsx'
XDB=pd.read_excel(nxl, sheet_name=2)

XDB.head(50)

XD= XDB[['Business','Age','Female','Years','Certificates','Feedback','Salary','NPS']]
yd=XDB[['College']]
```

2. Implementaci√≥n del modelo

```python
mar=DecisionTreeClassifier(criterion='gini', max_depth=8) #aqu√≠ le dejimos: trabaja con el indice de gini y con las 4 variables
mar.fit(XD,yd)
#mar= modelo de arbol
#max_depth: variables disponibles, profundidad (trabajar con este o menor n√∫mero de varibales )
#gini: es el criterio que indica el grado de error del modelo en la clasificaci√≥n de solicitantes

#SE PROCEDE CON EL DESPLIEGUE DEL √ÅRBOL

from sklearn.tree import export_graphviz #exportar
from pydotplus import graph_from_dot_data #crear el gr√°fico


vs= ['Business','Age','Female','Years','Certificates','Feedback','Salary','NPS']

dot_data=export_graphviz(mar, feature_names=vs) #esto es para construir el arbol
graph=graph_from_dot_data(dot_data)
graph.write_png('1.Arbolcollege.png')
```

3. Medir el desempe√±o

```python
ydp=mar.predict(XD)
cm=confusion_matrix(yd, ydp)
print(cm)

VN=cm[0,0];FP=cm[0,1]; FN=[1,0]; VP=[1,1]


Exactitud=(VP+VN)/(VP+VN+FP+FN) # funcionamiento general del modelo
print( 'La Exactitud es',Exactitud)

Sensibilidad= VP/(VN+FN)
print('La Sensibilidad es',Sensibilidad)

Especificidad=VN/(VN+FP)  # como se comport√≥ el modelo con los negados
print('La Especificidad es',Especificidad)

Precisi√≥n=VP/(VP+FP) # cual es el acierto clasificando los positivos
print('La Precisi√≥n es',Precisi√≥n)


TasaError=(FP+FN)/(VP+VN+FP+FN) #equivocaci√≥n general
print('La Tasa de Error es', TasaError)

PredNC=(VN+FN)/(VN+FP+FN+VP) #como funciona el modelo negando cr√©ditos
print('La tasa de predicci√≥n negativa es',PredNC)


```

```python
XDB2=pd.read_excel(nxl, sheet_name=2)
XDB2.head(100)

yd2=XDB2[['College']]
yd2.head(10)

#Donde est√°n los que no poseen informaci√≥n
filas=np.where((np.isnan(yd2))==True)[0]
filas
print('las filas que no poseen informaci√≥n son')
print(filas)



#Predicci√≥n

XD2 = XDB2[['Business', 'Age', 'Female', 'Years', 'Certificates', 'Feedback', 'Salary', 'NPS']]

ydp2 = mar.predict(XD2)
print("Las predicciones son", ydp2)



#Informe
dfp = pd.DataFrame(np.column_stack((XD2, ydp2)))
dfp.columns = ['Business', 'Age', 'Female', 'Years', 'Certificates', 'Feedback', 'Salary', 'NPS', 'Predicci√≥n']

print(dfp.head())


dfp.to_excel('InformeCollege.xlsx', index=False)



print('Los negados pronosticados son:', len(np.where(ydp2 == 0)[0]))
print('Los aprobados pronosticados son:', len(np.where(ydp2 == 1)[0]))
```
